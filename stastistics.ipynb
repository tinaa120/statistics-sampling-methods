## 3. Stratified Sampling:

code for this

for example : Stratify by 'Gender'

```python

n_samples_per_stratum = 5


stratified_sample = (
    csv_1.groupby('Gender', group_keys=False)
    .apply(lambda x: x.sample(n=min(n_samples_per_stratum, len(x)), random_state=1))
    .reset_index(drop=True)  # Optionally reset the index to avoid group keys issues
)

print(stratified_sample)
```

## 4.CLUSTER SAMPLING

Code for this :
<br> Cluster Definition: Each unique hospital in the dataset represents a distinct cluster.
<br>  Random Selection: A random sample of hospitals (clusters) is chosen to form the final sample.
<br>  Complete Inclusion: All individuals from the selected hospitals are included in the sample, ensuring comprehensive representation from the sampled clusters.

```python

import numpy as np 

num_clusters = 20  


clusters = csv_1['Hospital'].unique()
sampled_clusters = np.random.choice(clusters, num_clusters, replace=False)  
cluster_sample = csv_1[csv_1['Hospital'].isin(sampled_clusters)]  

print(cluster_sample)
```

## 3.SNOWBALL SAMPLING

Code for this

 letâ€™s assume we are starting with individuals who have a specific condition or are from a specific hospital.
Here, we simulate the snowball sampling process. We'll assume that each participant in the sample can refer up to a specified number of additional participants.

```python

print(csv_1.head())
print(csv_1.columns)

initial_condition = 'Diabetes'
initial_sample_size = 5

initial_sample = csv_1[csv_1['Medical Condition'] == initial_condition].head(initial_sample_size)

print("Initial Sample:")
print(initial_sample)

referral_limit = 10

all_sampled = set(initial_sample.index)

participants_to_process = list(initial_sample.index)

def refer_additional_participants(participant_indices, referral_limit):
    new_samples = set()
    for idx in participant_indices:
        participant_hospital = csv_1.loc[idx, 'Hospital']
        referrals = csv_1[csv_1['Hospital'] == participant_hospital]
        referrals = referrals[~referrals.index.isin(all_sampled)]
        new_samples.update(referrals.head(referral_limit).index)
    return new_samples

while participants_to_process:
    new_referrals = refer_additional_participants(participants_to_process, referral_limit)
    new_participants = new_referrals - all_sampled
    if not new_participants:
        break
    all_sampled.update(new_participants)
    participants_to_process = list(new_participants)

final_sample = csv_1.loc[list(all_sampled)]

print("Final Sample:")
print(final_sample)
```
